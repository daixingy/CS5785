```python
'''
    read data from iris.data
'''
%matplotlib  inline

import re

file = open("./dataset/iris.data")
lines = file.readlines()
data = []
for line in lines:
    line = re.split(',|\n', line)
    data.append(line[:-1])

data = data[:-1]
for i in range(len(data)):
    print(data[i])
```

```python
'''
    transfer string into data
'''
for i in range(len(data)):
    for j in range(len(data[i])-1):
        data[i][j] = float(data[i][j])

for i in range(len(data)):
    print(data[i])
```
